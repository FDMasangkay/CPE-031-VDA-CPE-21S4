```python
import matplotlib.pyplot as plt
import pandas as pd


health_status_counts = data2['GeneralHealth'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(health_status_counts.index, health_status_counts.values)
plt.xlabel("General Health Status")
plt.ylabel("Number of Individuals")
plt.title("Distribution of Individuals across General Health Statuses")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

'''
Question: What does the distribution of individuals across general health implies?
- By analyzing the chart, we can say that most of the individuals in this dataset report responded that they have either 'Very Good' and 'Good' health status.
Then followed by 'Fair' and 'Poor' health status in which does not cater half of the individuals.

'''
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# List of vaccine columns
vaccine_columns = ['PneumoVaxEver', 'FluVaxLast12']

data2[vaccine_columns] = data2[vaccine_columns].fillna('No')

# Create a figure for the plot
plt.figure(figsize=(12, 6))

for vaccine in vaccine_columns:
    vaccine_status_counts = data2[vaccine].value_counts().sort_index()

    vaccine_status_counts.plot(kind='bar', stacked=True, width=0.8, label=vaccine)

# Formatting the plot
plt.xlabel("Vaccine Status")
plt.ylabel("Number of Individuals")
plt.title("Distribution of Vaccine Status")
plt.xticks(rotation=0)
plt.legend(title="Vaccine")
plt.tight_layout()

# Show the plot
plt.show()

'''
Question: Does the respondents get their vaccine? How does it correlate to general health?
- The other graph shows that the respondents has very good and good health status, while majority of them does not get their vaccine.
We can say that the vaccine does make an individual gain a better health in general.

'''
```
